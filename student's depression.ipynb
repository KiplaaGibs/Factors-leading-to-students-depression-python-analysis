```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Gender', data=df, palette='coolwarm')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Age Distribution
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'], bins=10, kde=True, color='darkblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#correlation
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#level of academic pressure to depression 
```

```python
df['Pressure_Level'] = pd.cut(df['Academic Pressure'], bins=3, labels=['Low', 'Medium', 'High'])
sns.barplot(x='Pressure_Level', y='Depression', data=df)
```

```python
#family history of MI and depression 
```

```python
from scipy.stats import ttest_ind
group_yes = df[df['Family History of Mental Illness'] == 'Yes']['Depression']
group_no = df[df['Family History of Mental Illness'] == 'No']['Depression']
ttest_ind(group_yes, group_no)
```

```python
from sklearn.linear_model import LinearRegression
X = df[['Academic Pressure', 'CGPA', 'Work Pressure']]
y = df['Depression']
model = LinearRegression().fit(X, y)
print(model.coef_)
```
